```python
import pandas as pd
```

```python
fileName = 'Vehicle Check\cCalRecycle_NorthBranch_DataManagerTicketExport.xlsx'
```

# This opens the excell we need. Has columns preselected. Can change this

```python
df = pd.read_excel(fileName, usecols = ['Zone Name', 'Addr No', 'Addr St', 'Service Code', 'Is Void', 'Unit Count', 'End Time',
                                       'Ticket Number', 'Disposal Monitor Name', 'Ticket Notes'], index_col= "Zone Name") #index_col = "Zone Name"
```

# This cell will rearrage our columns in any order we want

```python
df = df[['Addr No', 'Addr St', 'Service Code', 'Is Void', 'Unit Count', 'End Time',
         'Ticket Number', 'Disposal Monitor Name', 'Ticket Notes']]
```

# This will take out all special Chars in RTR service code cells

```python
spec_chars = ["!",'"',"#","%","&","'","(",")",
              "*","+",",","-",".","/",":",";","<",
              "=",">","?","@","[","\\","]","^","_",
              "`","{","|","}","~","â€“"]
for char in spec_chars:
    df["Service Code"] = df["Service Code"].str.replace(char, ' ')
    
df["Service Code"] = df["Service Code"].str.split().str.join(" ")

codes = ["CO4", "4"]

df = df[(df["Is Void"] != True) & (df["Service Code"].isin(codes))]

# df.loc["Zone Name"]
df
    
```
